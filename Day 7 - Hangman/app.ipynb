```python
# Get word length
def get_word_length():
    word_length = None
    while word_length is None:
        try:
            _input = int(input("Please enter the length of the word you want to guess (3-10): \n"))
            if _input < 3 or _input > 10:
                raise Exception()
            word_length = _input
        except:
            print("Invalid word length, try again.")
    return word_length
        
# Get random word of length word_length
import requests
import random

def get_word():
    word_site = "https://www.mit.edu/~ecprice/wordlist.10000"
    response = requests.get(word_site)
    words = response.content.splitlines()
    word_length = get_word_length()
    filtered_words = [word for word in words if len(word) == word_length]
    if not filtered_words:
        return None
    word = random.choice(filtered_words).decode("utf-8")
    return word

def get_rendered_word():
    output = ["_" for _ in range(len(word))]
    word_letters = list(word)
    for _index, _  in enumerate(output):
        for letter in visited_letters:
            if letter == word_letters[_index]:
                output[_index] = letter
    output_word = " ".join(output)
    return output_word

def is_word_done():
    output_word = get_rendered_word()
    if output_word == " ".join(word):
        return True
    return False

# Global
lives = 6
visited_letters = []
word = get_word()


# Render logic
def render():
    global lives
    output_word = get_rendered_word()
    print(f"{output_word}\t|\tVisited letters: {visited_letters}\t|\tLives: {lives}")

# Get letter
def get_letter():
    global visited_letters
    letter = None
    while letter is None:
        try:
            _input = str(input("Please enter a letter: \n")).lower()
            if _input in visited_letters or len(_input) != 1 or not _input.isalpha():
                raise Exception()
            letter = _input
            visited_letters.append(letter)
        except:
            print("Invalid or used letter, try again.")
    return letter

# Update game state
def update():
    global lives
    if visited_letters[-1] not in word:
        lives -= 1

# Game
while lives > 0:
    render()
    get_letter()
    update()
    if is_word_done():
        print(f"Congrats! You found the word {word} with {lives} lives left.")
        break
else:
    print(f"Game Over, word was {word}.")
```
